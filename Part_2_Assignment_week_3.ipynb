```python
# nlp_spacy_reviews.py

import spacy
from textblob import TextBlob
import matplotlib.pyplot as plt

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample Amazon Product Reviews
reviews = [
    "I love the Samsung Galaxy phone. The display is stunning!",
    "The Nike running shoes are very comfortable and stylish.",
    "Apple MacBook is too expensive but performs well.",
    "I am disappointed with the Lenovo battery life.",
    "Sony headphones have excellent sound quality.",
    "The new Dell laptop overheats quickly and crashes often.",
    "Canon cameras are the best for photography.",
    "The quality of this Xiaomi phone is terrible.",
    "I highly recommend the ASUS gaming monitor.",
    "Avoid the Huawei tablet – it's too slow and buggy."
]

# Collect results
entities = []
sentiment_summary = {"Positive": 0, "Negative": 0, "Neutral": 0}

# Process each review
for review in reviews:
    print("\nReview:", review)

    # spaCy NER
    doc = nlp(review)
    print("Named Entities:")
    for ent in doc.ents:
        print(f" - {ent.text} ({ent.label_})")
        entities.append((ent.text, ent.label_))

    # Sentiment Analysis with TextBlob
    sentiment = TextBlob(review).sentiment.polarity
    if sentiment > 0:
        label = "Positive"
    elif sentiment < 0:
        label = "Negative"
    else:
        label = "Neutral"
    sentiment_summary[label] += 1
    print(f"Sentiment: {label} (Polarity: {sentiment:.2f})")

# Plot sentiment summary
labels = sentiment_summary.keys()
values = sentiment_summary.values()

plt.bar(labels, values, color=["green", "red", "gray"])
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("sentiment_distribution.png")
plt.show()
```

Ethics & Optimization
✅ 1. Identify Potential Biases in MNIST or Amazon Reviews Models
a. MNIST Dataset Biases
Bias Source: The MNIST dataset contains only grayscale, handwritten digits from Western-style numerals.

Potential Bias:

Cultural bias: It doesn’t account for different numeral systems (e.g., Arabic, Hindi).

Overfitting to neat, centered digits due to training data uniformity.

b. Amazon Reviews Model Biases
Bias Source: Reviews and ratings from real users can reflect societal biases (e.g., gender, race, language use).

Potential Bias:

Sentiment classifiers may misinterpret sarcasm or non-standard English.

Disproportionate representation of product categories or reviewers.

Association of certain words with negative/positive sentiment unfairly (e.g., gendered language).

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import spacy
from spacy.matcher import Matcher

print("========== MNIST Bias Detection ==========")

# Load and visualize MNIST
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Display digit samples
plt.figure(figsize=(10, 1))
for i in range(10):
    plt.subplot(1, 10, i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.axis('off')
plt.suptitle("Sample MNIST Digits")
plt.show()

# Digit distribution
unique, counts = np.unique(y_train, return_counts=True)
dist = dict(zip(unique, counts))
print("Digit Distribution:", dist)

print("\n========== Amazon Tweet Bias Detection ==========")

# Load tweets dataset
df = pd.read_csv("https://raw.githubusercontent.com/dD2405/Twitter_Sentiment_Analysis/master/train.csv")

bias_keywords = ['woman', 'man', 'girl', 'boy', 'black', 'white']
df['bias_flag'] = df['tweet'].apply(lambda x: any(word in x.lower() for word in bias_keywords))

biased_tweets = df[df['bias_flag']]
print("Biased Tweets Sample:")
print(biased_tweets[['tweet']].head())

print("\n========== Mitigation Using spaCy ==========")

nlp = spacy.load("en_core_web_sm")
matcher = Matcher(nlp.vocab)
pattern = [{"LOWER": "angry"}, {"LOWER": "woman"}]
matcher.add("BIAS_PATTERN", [pattern])

doc = nlp("The angry woman was not happy with the service.")
matches = matcher(doc)

if matches:
    for _, start, end in matches:
        span = doc[start:end]
        print("Matched Bias Phrase:", span.text)
else:
    print("No biased phrase found.")

print("\n========== Training MNIST CNN Model ==========")

# Normalize and reshape
x_train = x_train / 255.0
x_test = x_test / 255.0
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Build CNN
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

# Compile & Train
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=3, validation_data=(x_test, y_test))

# Save model
model.save("mnist_model.h5")
print("✅ Model saved as mnist_model.h5")
```
